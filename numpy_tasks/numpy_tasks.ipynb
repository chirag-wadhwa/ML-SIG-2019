# Question 2:
### Write a function that performs min-max scaling on a numpy array across rows 

## Example:

### Input 
$\begin{bmatrix}1& 2& 3\\10 & 15 & 20 \\ -5 & -11&  9\end{bmatrix}$
    
### Output
$\begin{bmatrix} 0 & 0.5 & 1 \\ 0 & 0.5 & 1 \\ 0.3 & 0 & 1 \end{bmatrix}$


```python
# Answer 2 - using for loops
def scaling_with_loops(arr):
    sh = arr.shape
    li = []
    for row in arr:
        temp = row.max() - row.min()
        for ele in row:
            temp_ele = (ele - row.min())/temp
            li.append(temp_ele)
    arr = np.array(li).reshape(sh)
    return arr
```

```python
# Answer 2 - without using for loops
def scaling_without_loops(arr):
    sh = arr.shape
    min_row = arr.min(axis=1)
    max_row = arr.max(axis=1)
    min_row = np.repeat(min_row, sh[1]).reshape(sh)
    max_row = np.repeat(max_row, sh[1]).reshape(sh)
    diff = max_row - min_row
    arr = (arr - min_row)/diff
    return arr
```

val = 5

# Question 4:
### Implement A sliding window average of a given array

Example: https://www.youtube.com/watch?v=ZoaEDbivmOE

```python
# Answer 4 - using for loops
import numpy as np
import matplotlib.pyplot as plt
arr = np.zeros((27,27))
arr[3:24,4:7]=255
arr[3:6,4:20]=255
arr[12:15,4:20]=255
arr[3:15,17:20]=255
plt.imshow(arr, 'jet')
```

```python
for i in range(1,arr.shape[0]-1):
    for j in range(1,arr.shape[1]-1):
        sum = 0
        for k in range(i-1,i+2):
            for l in range(j-1,j+2):
                sum = sum+arr[k,l]
                arr[i,j] = sum/9
plt.imshow(arr, 'jet')
```

### Implement John Conway's Game of Life

You can read about the rules of the game and play it at: https://bitstorm.org/gameoflife/

```python
# Answer 5
import numpy as np
import matplotlib.pyplot as plt
arr = np.zeros((100,100))
arr[40:60,30:70] = 255
arr[45:55,35:45] = 0
arr[45:55,55:65] = 0
plt.imshow(arr, 'gray')
```

```python
for i in range(1,arr.shape[0]-1):
    for j in range(1,arr.shape[1]-1):
        if arr[i,j] == 0:
            temp = 0
            for k in range(i-1,i+2):
                for l in range(j-1,j+2):
                    if arr[k,l] == 255:
                        temp += 1
                    if temp == 3:
                        arr[i,j] = 255
                        break
        elif arr[i,j] == 255:
            temp = 0
            for k in range(i-1,i+2):
                for l in range(j-1,j+2):
                    if arr[k,l] == 255:
                        temp += 1
            if temp < 2:
                arr[i,j] = 0
            elif temp>3:
                arr[i,j] = 0
plt.imshow(arr, 'gray')
```
